# 350 samples FedFor hugadb

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm


save_dir = "/..../results/c2fedfor_"
noise_std_values = [0.1 , 0.3 , 0.5 , 0.7 ,1 , 2] 
corrupted_clients_counts = [3 , 4,6 , 8  , 9]


results = {noise: [] for noise in noise_std_values}


for noise in noise_std_values:
    for cc in corrupted_clients_counts:
        filename = f"Nash_Equilibrium_Details_noise_{noise}_c{cc}.csv"
        file_path = os.path.join(save_dir, filename)
        if os.path.exists(file_path):
            df = pd.read_csv(file_path)

            occurrence_count = (df['Combination'] == 1111111111 ).sum()
            results[noise].append(occurrence_count)
        else:
            print(f"File {file_path} does not exist.")
            results[noise].append(0)


from matplotlib.colors import ListedColormap


custom_colors = ['blue', 'orange', 'green', 'red', 'purple', 'brown']


cmap = ListedColormap(custom_colors)
markers = ['x','o','s','d', '*','v', '^', '<', '>','h', 'D']
linestyles = ['-', '-.' , ':']

plt.figure(figsize=(10, 6))
for i, noise in enumerate(noise_std_values):
    color = cmap(i) 
    marker = markers[i % len(markers)]
    linestyle = linestyles[i % len(linestyles)]
    
    plt.plot(corrupted_clients_counts, results[noise],
             marker=marker, linestyle=linestyle, color=color,
             label=f'Noise Std = {noise}')

plt.xlabel("Number of Low Quality Clients")
plt.ylabel("Occurrences of all clients coalition")
plt.legend(title="Noise Std")
plt.title('FedFor on HuGaDB')
plt.grid(True)
plt.show()
```

# 350 samples FedLR spambase

```python
import pandas as pd
import matplotlib.pyplot as plt


try:
    df = pd.read_csv('/..../results/c1fedlr_/nash_occurrence_results.csv', sep=None, engine='python')
except:
    df = pd.read_csv('/..../results/c1fedlr_/nash_occurrence_results.csv', delimiter=';')


df.columns = df.columns.str.strip()  
df = df.rename(columns={df.columns[0]: "num_bad_clients"}) 


if "num_bad_clients" not in df.columns:
    raise KeyError("Column 'num_bad_clients' not found in DataFrame!")


noise_levels = [0.1, 0.3, 0.5, 0.7, 1, 2]


markers = ['x', 'o', 's', 'd', '*', 'v', '^', '<', '>', 'h', 'D']
linestyles = ['-', '-.']

plt.figure(figsize=(8, 5))


x = df['num_bad_clients']


for i, nl in enumerate(noise_levels):
    col_name = str(float(nl)) 
    if col_name not in df.columns:
        print(f"Warning: Column '{col_name}' not found in DataFrame!")
        continue  

    y = df[col_name]
    marker = markers[i % len(markers)]
    linestyle = linestyles[i % len(linestyles)]
    plt.plot(x, y, marker=marker, linestyle=linestyle, label=f'Noise Std = {nl}')


plt.xlabel('Number of Low Quality Clients')
plt.ylabel('Occurrences of all clients coalition')
plt.title('FedLR on Spambase')
plt.grid(True)
plt.legend()
plt.show()
```
